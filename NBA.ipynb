```python
import requests
import pandas as pd

# Below is the code for scrapping " https://www.nba.com/standings " for the year 2020-21

url= "https://stats.nba.com/stats/leaguestandingsv3"

# The Website im trying to scrape -
# https://www.nba.com/standings

# Get the Request header from the page you want to scrape
headers = {
    'Accept': 'application/json, text/plain, */*',
    'Accept-Encoding': 'gzip, deflate, br',
    'Accept-Language': 'en-GB,en-US;q=0.9,en;q=0.8',
    'Cache-Control': 'no-cache',
    'Connection': 'keep-alive',
    'Host': 'stats.nba.com',
    'Origin': 'https://www.nba.com',
    'Pragma': 'no-cache',
    'Referer': 'https://www.nba.com/',
    'sec-ch-ua-mobile': '?0',
    'Sec-Fetch-Dest': 'empty',
    'Sec-Fetch-Mode': 'cors',
    'Sec-Fetch-Site': 'same-site',
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.82 Safari/537.36',
    'x-nba-stats-origin': 'stats',
    'x-nba-stats-token': 'true'
}


# Get the Query String Parameter from the page you want to scrape
params = (
("LeagueID", "00"),
("Season", "2020-21"),
("SeasonType", "Regular Season"))

# Use the requests python package to scrape the page
response = requests.get(url, headers=headers, params=params)

# If everything goes right you will see "<Response [200]>" when you print the response
print(response)

# Web request into JSON format
jsn = response.json()

# Reformat the JSON to DataFrame ( You can later save it as a CSV )
frame1 = pd.DataFrame(jsn['resultSets'][0]['rowSet'])

# Assigning column headers
frame1.columns = jsn['resultSets'][0]['headers']

# print the top 10 records
print(frame1.head(10))


#save it to excel 
#frame.to_excel(location, sheet_name = 'Standings',engine='xlsxwriter')
#frame.to_excel(location, sheet_name='Standings',index=False)
```

```python
# Below is the code for scrapping " https://www.nba.com/standings " for the year 2019-20

params = (
("LeagueID", "00"),
("Season", "2019-20"),
("SeasonType", "Regular Season"))


response = requests.get(url, headers=headers, params=params)

# If everything goes right you will see "<Response [200]>" when you print the response
print(response)

# Web request into JSON format
jsn = response.json()

# Reformat the JSON to DataFrame ( You can later save it as a CSV )
frame2 = pd.DataFrame(jsn['resultSets'][0]['rowSet'])

# Assigning column headers
frame2.columns = jsn['resultSets'][0]['headers']

# print the top 10 records
print(frame2.head())

#save it to excel 
location = "/content/drive/MyDrive/NBA Scraper Code/2019-20 NBA Season.xlsx"
frame2.to_excel(location, sheet_name='Standings',index=False)
```

```python
# Below is the code for scrapping " https://www.nba.com/stats/teams/advanced/ " for the year 2020-21



url= "https://stats.nba.com/stats/leaguedashteamstats"
# The Website im trying to scrape -
# https://www.nba.com/stats/teams/advanced/


params = (
('Conference', ''),
('DateFrom', ''),
('DateTo', ''),
('Division', ''),
('GameScope', ''),
('GameSegment', ''),
('LastNGames', '0'),
('LeagueID', '00'),
('Location',''),
('MeasureType', 'Advanced'),
('Month', '0'),
('OpponentTeamID', '0'),
('Outcome',''),
('PORound', '0'),
('PaceAdjust', 'N'),
('PerMode', 'PerGame'),
('Period', '0'),
('PlayerExperience', ''),
('PlayerPosition', ''),
('PlusMinus', 'N'),
('Rank', 'N'),
('Season', '2020-21'),
('SeasonSegment', ''),
('SeasonType', 'Regular Season'),
('ShotClockRange', ''),
('StarterBench', ''),
('TeamID', '0'),
('TwoWay', '0'),
('VsConference', ''),
('VsDivision', ''))

response = requests.get(url, headers=headers, params=params)

# If everything goes right you will see "<Response [200]>" when you print the response
print(response)

# Web request into JSON format
jsn = response.json()

# Reformat the JSON to DataFrame ( You can later save it as a CSV )
frame3 = pd.DataFrame(jsn['resultSets'][0]['rowSet'])

# Assigning column headers
frame3.columns = jsn['resultSets'][0]['headers']

# print the top 10 records
print(frame3.head())
```

```python
#save it to excel 

location = "/content/drive/MyDrive/NBA Scraper Code/NBA Stats.xlsx"

writer = pd.ExcelWriter(location, engine='xlsxwriter')

frame1.to_excel(writer, sheet_name='Standing 2020-21',index=False)
frame2.to_excel(writer, sheet_name='Standing 2019-20',index=False)
frame3.to_excel(writer, sheet_name='Advansed Stats 2020-21',index=False)

writer.save()
```
